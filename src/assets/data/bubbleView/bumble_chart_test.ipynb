```python
import json
import pandas as pd
import plotly.graph_objects as go
import plotly.express as px
from sklearn.manifold import TSNE
from sklearn.cluster import KMeans
import numpy as np
import matplotlib.colors as mcolors
import yake

import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ersetzen Sie 'ihre_daten.json' mit dem Pfad zu Ihrer JSON-Datei
with open('./test.json', 'r',encoding='utf-8') as file:
    liste = json.load(file)
data = [item for sublist in liste for item in sublist]
#df_topic = pd.json_normalize(data)
df_topics = pd.DataFrame(data)
```

```python
df_topics
```

```python

# YAKE! Keyword-Extraktion initialisieren
kw_extractor = yake.KeywordExtractor(lan="de", n=1, dedupLim=0.9, top=1, features=None)

# Funktion zum Extrahieren des Top-Keywords als Oberbegriff
def extract_keyword(text):
    keywords = kw_extractor.extract_keywords(text)
    if keywords:
        # Nur das Top-Keyword zurückgeben
        return keywords[0][0]
    else:
        return "Allgemein"

# Anwenden der Funktion auf Ihre Themen
df_topics['short_name'] = df_topics['topic'].apply(extract_keyword)
```

```python
df_topics
```

```python

# Umwandeln der Liste von Listen in ein NumPy-Array
embeddings_list = list(df_topics['embedding'])
embeddings_array = np.array(embeddings_list)


# Dimensionalitätsreduktion
tsne = TSNE(n_components=2, random_state=42, perplexity=min(17 - 1, 30))  # 17 ist die Anzahl der Samples, 30 ist die Standard-Perplexität
embeddings = tsne.fit_transform(embeddings_array)



# Fügen Sie die reduzierten Dimensionen zu df_topics hinzu
df_topics['x'] = embeddings[:, 0]
df_topics['y'] = embeddings[:, 1]

#colors = [mcolors.to_hex(c) for c in colors]
# Bestimmen Sie die Anzahl der Gruppen
num_clusters = 10  # oder eine andere Anzahl, die für Ihre Daten geeignet ist

# Führen Sie k-means auf den reduzierten Dimensionen aus
kmeans = KMeans(n_clusters=num_clusters, random_state=42)
df_topics['group'] = kmeans.fit_predict(df_topics[['x', 'y']])

# Generieren eines Farbschemas basierend auf der Anzahl der Gruppen
cmap = plt.get_cmap('viridis', num_clusters)

# Weisen Sie jeder Gruppe eine Farbe zu und konvertieren Sie diese in Hex-Code
group_to_color = {group: mcolors.to_hex(cmap(i)) for i, group in enumerate(range(num_clusters))}
df_topics['color'] = df_topics['group'].map(group_to_color)

df_topics
```

```python
# Erstellen des Bubble-Charts mit den verkürzten Namen
fig = px.scatter(df_topics, x='x', y='y',
                 size='size',  # Größe der Blasen
                 color='color',
                 text='short_name',  # Textlabel der Blasen sind jetzt die Kurzformen
                 hover_name='topic',  # Vollständiger Name beim Hover
                 size_max=60  # Maximalgröße für Blasen
                 )

# Farbskala entfernen
fig.update_layout(coloraxis_showscale=False, showlegend = False)

start_date = "2022-01-01"
end_date = "2023-01-01"

# Titel und Achsenbeschriftungen
fig.update_layout(
    title=f"Top {num_clusters} Themen Bundestag zwischen {start_date} und {end_date}",
    xaxis_title="",
    yaxis_title="",
    xaxis=dict(showgrid=False, zeroline=False, visible=False),  # Achsen unsichtbar machen
    yaxis=dict(showgrid=False, zeroline=False, visible=False)
)



# Erstellen Sie einen String für die Annotation
time_period_annotation = f"Von {start_date} bis {end_date}"

# Annotation für den Zeitraum
fig.add_annotation(
    x=0.5,
    y=-0.1,
    xref='paper',
    yref='paper',
    text=time_period_annotation,
    showarrow=False,
    font=dict(
        size=12
    )
)

# Pfeil hinzufügen
fig.add_shape(
    type='line',
    x0=0.2,  # Anfang des Pfeils (relativ zur Zeichnungsfläche)
    y0=-0.05,  # Vertikale Position des Pfeils
    x1=0.8,  # Ende des Pfeils (relativ zur Zeichnungsfläche)
    y1=-0.05,  # Vertikale Position des Pfeils (gleich wie x0, da gerader Pfeil)
    line=dict(
        color='black',
        width=2
    ),
    xref='paper',
    yref='paper'
)
# Pfeilspitze hinzufügen
fig.add_shape(
    type='line',
    x0=0.8,  # Anfang der Pfeilspitze
    y0=-0.05,  # Vertikale Position der Pfeilspitze
    x1=0.79,  # Ende der Pfeilspitze (leicht links vom Ende des Pfeils)
    y1=-0.055,  # Vertikale Position der Pfeilspitze (leicht nach unten)
    line=dict(
        color='black',
        width=2
    ),
    xref='paper',
    yref='paper'
)

fig.add_shape(
    type='line',
    x0=0.8,  # Anfang der Pfeilspitze (wie beim ersten Pfeilteil)
    y0=-0.05,  # Vertikale Position der Pfeilspitze (wie beim ersten Pfeilteil)
    x1=0.79,  # Ende der Pfeilspitze (leicht links vom Ende des Pfeils)
    y1=-0.045,  # Vertikale Position der Pfeilspitze (leicht nach oben)
    line=dict(
        color='black',
        width=2
    ),
    xref='paper',
    yref='paper'
)



# Texte der Themen immer anzeigen
fig.update_traces(textposition='middle center')

fig.show()
```
